```python


import pandas as pd
import numpy as np
import plotly.express as px
from textblob import TextBlob
df = pd.read_csv('netflix_titles.csv')
```

```python
df.shape
```

```python



print(df.isnull().sum())


df.dropna(subset=['director', 'cast'], inplace=True)

df['rating'].fillna('Not Rated', inplace=True)

df.drop_duplicates(inplace=True)

df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

df['country'] = df['country'].str.strip()
df.head()
```

```python
x = df.groupby(['rating']).size().reset_index(name ='count')
print(x)
```

```python
piechar = px.pie(x, values='count', names = 'rating', title='Distribution of content rating on netfix')
piechar.show()
```

```python
director_list = pd.DataFrame()
```

```python
director_list = df['director'].str.split(',', expand=True).stack()
director_list = director_list.reset_index(drop=True)
director_list = director_list.to_frame(name='Director')
director_list
```

```python
directors = director_list.groupby(['Director']).size().reset_index(name='Total count').sort_values(by=['Total count'], ascending=False)
directors
```

```python
directors = directors.sort_values(by=['Total count'], ascending=True)
directors
```

```python
directors_top10 = directors[0:10]
directors_top10
```

```python
bar = px.bar(directors_top10, x='Total count', y='Director', title='Top 10 Directors')
bar.show()
```

```python
cast_df =pd.DataFrame()
cast_df = df['cast'].str.split(',', expand = True).stack()
cast_df = cast_df.to_frame()
cast_df.columns = ['Actors']
actor = cast_df.groupby(['Actors']).size().reset_index(name='Total count')
actor.sort_values(by=['Total count'], ascending=False)
top_ten_actor = actor[0:10]
```

```python
bar_chart = px.bar(top_ten_actor, x ='Total count', y='Actors', title ='Top_ten_actor')
bar_chart.show()
```

Analysis content produced on netflix based on years


```python
df1 = df[['type', 'release_year']]
df1 = df1.rename(columns={"relase_year": "release_year", "type": "Type"})
df2 = df1.groupby(['release_year', 'Type']).size().reset_index(name='Count')
df2.sort_values(by=['release_year'], ascending=True)
```

```python
df2 = df2[df2['release_year']>2000]
graph = px.line(df2, x='release_year', y='Count', color ='Type', title='content produced on netflix base on year')
graph.show()
```

Sentiment Analysis of Netfilx content

```python
df3 = df[['release_year', 'description'] ]
df3 = df3.rename(columns={"relase_year":"release_year", 'description':'description'})
for index, row in df3.iterrows():
  d=row['description']
  testimonial = TextBlob(d)
  p = testimonial.sentiment.polarity
  if p==0:
    sent ='Neutral'
  elif p>0:
    sent = 'postive'
  else:
    sent = 'negative'
  df3.loc[[index, 2], 'sentiment'] = sent

df3 = df3.groupby(['release_year', 'sentiment']).size().reset_index(name='Count')
df3 = df3[df3['release_year']>2000]
bargraph = px.bar(df3, x='release_year', y='Count', color='sentiment', title='Sentiment Analysis')
bargraph.show()
```
